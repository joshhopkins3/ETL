```python
import pandas as pd
```

```python
# Create Url's to download files
sports = ['nfl','nba','nhl','mlb']
for sport in sports:
    print(sport)
    
    for i in range(10,22):
        
        if sport == 'mlb':
            url = f'https://www.sportsbookreviewsonline.com/scoresoddsarchives/{sport}/{sport}%20odds%2020{i}.xlsx'

        else:
            year = str(i)+"-"+str(i+1)
            url = f'https://www.sportsbookreviewsonline.com/scoresoddsarchives/{sport}/{sport}%20odds%2020{year}.xlsx'

        print(url)
```

```python
# MLB
sport = 'mlb'
mlb_df = pd.read_excel(f'ETL/Odds Data/{sport}/{sport} odds 2010.xlsx')
path = f'ETL/Odds Data/{sport}/{sport} odds 20'

for i in range(11,22):
    df2 = pd.read_excel(path+str(i)+'.xlsx')
    mlb_df.append(df2)    
```

```python
# NFL
sport = 'nfl'
nfl_df = pd.read_excel(f'ETL/Odds Data/{sport}/{sport} odds 2010-11.xlsx')
path = f'ETL/Odds Data/{sport}/{sport} odds 20'

for i in range(11,21):
    df2 = pd.read_excel(path+str(i)+"-"+str(i+1)+'.xlsx')
    nfl_df.append(df2) 
```

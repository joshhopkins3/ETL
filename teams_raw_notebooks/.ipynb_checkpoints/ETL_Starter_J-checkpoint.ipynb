# Clean NHL data

```python
nhl_df
```

```python
clean_nhl = nhl_df[['Date', 'VH', 'Team', 'Final','PuckLine', 'Year']].copy()
```

```python
clean_nhl
```

```python
def month(x):
    if len(str(x)) > 3:
        return int(str(x)[:2])
    else:
        return int(str(x)[:1])
    
def day(x):
    if len(str(x)) > 3:
        return int(str(x)[2:])
    else:
        return int(str(x)[1:])

    
clean_nhl["Month"] = clean_nhl['Date'].apply(lambda x: month(x))
clean_nhl["Day"] = clean_nhl['Date'].apply(lambda x: day(x))
```

```python
cols=['Year', 'Month', 'Day']
clean_nhl["Date"] = clean_nhl[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
```

```python
clean_nhl.head()
```

```python
clean_nhl = clean_nhl.drop(['Year', 'Month', 'Day'], axis=1)
```

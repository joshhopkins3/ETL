```python
#change Date to Datetime

def GrabMonth(x):
    if len(str(x)) > 3:
        return int(str(x)[:2])
    else:
        return int(str(x)[:1])
def GrabDay(x):
    if len(str(x)) >3:
        return int(str(x)[2:])
    else:
        return int(str(x)[1:])

 #Apply new defined functions to DF
new_nfl_df['Month'] = new_nfl_df['Date'].apply(lambda x: GrabMonth(x))
new_nfl_df['Day'] = new_nfl_df['Date'].apply(lambda x: GrabDay(x))

# Transform the Date column type
cols=['Year','Month','Day']
new_nfl_df['Date'] = new_nfl_df[cols].apply(lambda x: '/'.join(x.values.astype(str)), axis="columns")
new_nfl_df.drop((['Year','Month','Day']),axis=1,inplace=True)    
new_nfl_df['Date']=pd.to_datetime(new_nfl_df['Date'])

```

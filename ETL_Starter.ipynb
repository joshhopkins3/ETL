```python
'''
Brittany
'''

# NBA
sport = 'nba'
nba_df = pd.read_excel(f'Odds Data/{sport}/{sport} odds 2010-11.xlsx')
nba_df['Date'] = nba_df['Date'].apply(lambda x: str(2010) + str(x) if x > 900 else str(20110) + str(x))
path = f'Odds Data/{sport}/{sport} odds 20'

team_mapping = {
"Miami" : "Miami Heat",
"Boston": "Boston Celtics",
"Phoenix": "Phoenix Suns",
"Portland": "Portland Trail Blazers",
"Houston": "Houston Rockets",
"LALakers": "Los Angeles Lakers",
"Detroit": "Detroit Pistons",
"NewJersey": "New Jersey Devils",
"Cleveland": "Cleaveland Cavaliers",
"NewYork": "New York Knicks",
"Toronto": "Toronto Raptors",
"Philadelphia": "Philadelphia 76ers",
"Atlanta": "Atlanta Hawks",
"Memphis": "Memphis Grizzlies",
"Sacramento": "Sacramento Kings",
"Minnesota": "Minnesota Timberwolves",
"Milwaukee": "Milwaukee Bucks",
"NewOrleans": "New Orleans Pelicans",
"Chicago": "Chicago Bulls",
"OklahomaCity": "Oklahoma City Thunder",
"Charlotte": "Charlotte Hornets",
"Dallas": "Dallas Mavericks",
"Indiana": "Indiana Pacers",
"SanAntonio": "San Antonio Spurs",
"Utah": "Utah Jazz",
"Denver": "Denver Nuggets",
"GoldenState": "San Francisco Golden State Warriors",
"LAClippers": "Los Angeles Clippers",
"Washington": "Washington Wizards",
"Orlando": "Orlando Magic",
"Brooklyn": "Brooklyn Nets",
"Oklahoma City": "Oklahoma City Thunder",
"LA Clippers": "Los Angeles Clippers"
}

for i in range(11,21):
    df2 = pd.read_excel(path+str(i)+"-"+str(i+1)+'.xlsx')
    df2['Date'] = df2['Date'].apply(lambda x: str(20) + str(i) + str(x) if x > 900 else str(20) + str(i+1)+ str(0) + str(x))
    nba_df = nba_df.append(df2) 
    
nba_df=nba_df.drop(labels=['Rot','1st','2nd','3rd','ML','2H','4th'],axis=1)
nba_df["Team"]=nba_df["Team"].apply(lambda x: team_mapping[x])
nba_df['Date']=pd.to_datetime(nba_df['Date'], format='%Y%m%d')
```

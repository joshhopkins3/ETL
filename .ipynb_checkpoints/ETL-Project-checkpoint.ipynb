```python
'''
Josh
'''

# NHL
sport = 'nhl'
nhl_df = pd.read_excel(f'Odds Data/{sport}/{sport} odds 2010-11.xlsx')
path = f'Odds Data/{sport}/{sport} odds 20'

nhl_df['Year']= 2010

for i in range(11,20):
    df2 = pd.read_excel(path+str(i)+"-"+str(i+1)+'.xlsx')
    year = int('20'+str(i))
    df2['Year']= year
    nhl_df = nhl_df.append(df2)
    
nhl2 = pd.read_excel(path+'21.xlsx')
nhl2['Year'] = 2021
nhl_df = nhl_df.append(nhl2)
```

```python
# MLB
sport = 'mlb'
mlb_df = pd.read_excel(f'Odds Data/{sport}/{sport} odds 2010.xlsx')
path = f'Odds Data/{sport}/{sport} odds 20'

mlb_df['Year'] = 2010

for i in range(11,22):
    df2 = pd.read_excel(path+str(i)+'.xlsx')
    year = int('20'+str(i))
    df2['Year'] = year

    mlb_df = mlb_df.append(df2)
    
mlb_df.drop(axis = 1, columns = ['Pitcher','Rot', '1st', '2nd', '3rd', '4th',
   '5th', '6th', '7th', '8th', '9th','Open OU', 'Open',
   'Unnamed: 18', 'Close OU', 'Unnamed: 20','Run Line','Close OU', 'Unnamed: 22','RunLine','OpenOU','CloseOU'],inplace=True)
```

```python
# MLB DB Cleaning

'''
Build Function to Grab Month / Day of Month from current Date Column

'''
def GrabMonth(x):
    if len(str(x)) > 3:
        return int(str(x)[:2])
    else:
        return int(str(x)[:1])

def GrabDay(x):
    if len(str(x)) > 3:
        return int(str(x)[2:])
    else:
        return int(str(x)[1:])
    
# Apply new functions to DF
    
mlb_df['Month'] = mlb_df['Date'].apply(lambda x: GrabMonth(x))
mlb_df['Day'] = mlb_df['Date'].apply(lambda x: GrabDay(x))
```

```python
# Transform Date Column to DateTime

cols=['Year','Month','Day']
mlb_df['Date'] = mlb_df[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
mlb_df.drop(['Year','Month','Day'],axis=1,inplace=True)
mlb_df['Date'] = pd.to_datetime(mlb_df['Date'])

# Get Unique Team Names and Map to Updated Team names

teams = mlb_df['Team'].unique()
names = ['New York Yankees','Boston Red Sox','Philidelphia Phillies','Washington Nationals','Miami Marlins',
         'New York Mets','St Louis Cardinals','Cincinati Reds','Los Angeles Angels','Pitsburg Pirates',
         'Colorado Rockies', 'Milwaukee Brewers','Chicago Cubs','Atlanta Braves','San Diego Padres','Arizona Diamondbacks',
         'San Francisco Giants','Houston Astros','Cleveland Indians','Chicago White Sox','Toronto Blue Jays',
         'Texas Rangers','Detroit Tigers','Kansas City Royals','Minnesota Twins','Las Angeles Angels',
         'Seatle Mariners','Oakland Athletics','Baltimore Orioles','Tampa Bay Rays','Los Angeles Dodgers',
         'San Francisco Giants','Boston Red Sox', 'Chicago Cubs']

name_dict = dict(zip(teams,names))
mlb_df['Team'] = mlb_df['Team'].map(name_dict)

# Change Close to ML
mlb_df = mlb_df.rename({'Close':'ML'},axis=1)

# Drop NL Scores
mlb_df = mlb_df[mlb_df['Final']!='NL']
```
